```python
redshade_img =cv2.cvtColor(orginal_img,cv2.COLOR_BGR2RGB)#COLOR_BGR2RGB==> blue green red 2 red green blue color transfer 
```

```python
plt.imshow(redshade_img)
```

```python
blueshade_img =cv2.cvtColor(redshade_img,cv2.COLOR_RGB2BGR)# #orginal image
plt.imshow(blueshade_img)
#print(blueshade_img.shape)
```

```python
grey_img =cv2.cvtColor(blueshade_img,cv2.COLOR_BGR2GRAY)# which is the grey color image
plt.imshow(grey_img)
```

```python
print(grey_img)
```

```python
grey_img.shape
```

Lets Find the mean and average colour of the image

```python
avg_color_pre_row =np.average(grey_img,axis =0)
# finding the avg_per_row of the grey_img
#np.average()==> takes in an axis argument which finds the average across that axis=0
```

```python
avg_color =np.average(avg_color_pre_row,axis =0)
# finding the average across the average per row
```

```python
avg_color =np.uint8(avg_color)
print(avg_color)
```

```python
avg_color_org =np.average(orginal_img,axis =0)
avg_org_color =np.average(avg_color_org,axis= 0)
```

```python
avg_color1 =np.uint8(avg_org_color)
print(avg_color1)
```
